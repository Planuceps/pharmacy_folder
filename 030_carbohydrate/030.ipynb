```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('fermentation_results.csv')
```

```python
df
```

```python
df
df_transposed = df.transpose()

df_transposed.columns = ['Sample 1', 'Sample 2', 'Sample 3', 'Sample 4', 'Sample 5', 'Sample 6']
df_transposed
```

```python
df_transposed.columns
```

```python
df_transposed_a = df_transposed.drop('Sample 6', axis=1)
```

```python
df_transposed_a
```

```python
df_transposed_a.plot()
plt.xlabel('Time(min)')
plt.ylabel('Volume(ml)')
plt.title('Relation between Time and Volume for samples 1-5')
plt.legend(title='Sample')
for col in df_transposed_a.columns:
    plt.scatter(df_transposed_a.index, df_transposed_a[col], color=plt.gca().lines[df_transposed_a.columns.get_loc(col)].get_color(), s=10, zorder=3)
plt.grid(True)
plt.show()

plt.savefig("buffer_capacity_lowess.png") 
```

```python
df_transposed_b = df_transposed.drop('Sample 1', axis=1)
df_transposed_b = df_transposed_b.drop('Sample 3', axis=1)
df_transposed_b = df_transposed_b.drop('Sample 4', axis=1)
df_transposed_b = df_transposed_b.drop('Sample 5', axis=1)
```
