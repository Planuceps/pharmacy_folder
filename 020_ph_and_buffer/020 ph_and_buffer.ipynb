```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('buffer_HCL_data.csv')
df2 = pd.read_csv('buffer_NAOH_data.csv')
```

```python
df['b(mol/L)'] = np.nan
df['Buffer capacity'] = np.nan
df['b(mol/L)'] = ((0.5 * df["Volume (mL)"]) / (50 + df["Volume (mL)"])).round(3)
df['Buffer capacity'] = abs(df['b(mol/L)'].diff() / df['pH'].diff().round(3)).round(3)

df["Volume (mL)"] = df["Volume (mL)"].astype(str) + " HCL"

df = df.drop(index=0).reset_index(drop=True)

df
```

```python
df.to_csv('HCL.csv', index=False)
```

```python
df2['b(mol/L)'] = np.nan
df2['Buffer capacity'] = np.nan
df2['b(mol/L)'] = ((0.5 * df2["Volume (mL)"]) / (50 + df2["Volume (mL)"])).round(3)
df2['Buffer capacity'] = abs(df2['b(mol/L)'].diff() / df2['pH'].diff().round(3)).round(3)

df2["Volume (mL)"] = df2["Volume (mL)"].astype(str) + " NaOH"

df2 = df2.drop(index=0).reset_index(drop=True)
df2
```

```python
df2.to_csv('NaOH.csv', index=False)
```

```python
df_merged = pd.concat([df, df2], ignore_index=True)
df_merged = df_merged.sort_values(by="pH").reset_index(drop=True)
df_merged
```

```python
df_merged.to_csv("df_merged.csv", index=False)
```

```python
plt.figure(figsize=(12, 7))
plt.plot(
    df_merged["pH"], 
    df_merged["Buffer capacity"], 
    marker='o', 
    linestyle='-', 
    color='royalblue', 
    linewidth=2, 
    markersize=8, 
    markerfacecolor='orange', 
    markeredgecolor='black'
)
plt.xlabel("pH", fontsize=16)
plt.ylabel("Buffer Capacity (mol/L·pH)", fontsize=16)
plt.title("Relationship between pH and Buffer Capacity", fontsize=18, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout(pad=2)
plt.xlim(df_merged["pH"].min() - 0.5, df_merged["pH"].max() + 0.5)  # เพิ่มช่องว่างแกน x
plt.ylim(bottom=0)  # เริ่มแกน y ที่ 0
plt.savefig("buffer_capacity_plot.png", dpi=300)
plt.show()
```

```python
import statsmodels.api as sm

x = df_merged["pH"]
y = df_merged["Buffer capacity"]

# LOWESS smoothing
lowess = sm.nonparametric.lowess
y_smoothed = lowess(y, x, frac=0.2)  # frac=0.2 can be adjusted for smoothness

plt.figure(figsize=(12, 7))
plt.scatter(x, y, color='orange', s=50, label='Data Points')


plt.plot(y_smoothed[:, 0], y_smoothed[:, 1], color='red', linewidth=3, label='LOWESS Smoothed')


plt.xlabel("pH", fontsize=16)
plt.ylabel("Buffer Capacity (mol/L·pH)", fontsize=16)
plt.title("Relationship between pH and Buffer Capacity (LOWESS)", fontsize=18, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout(pad=2)
plt.xlim(x.min() - 0.5, x.max() + 0.5)
plt.ylim(bottom=0)
plt.legend()
plt.savefig("buffer_capacity_lowess.png", dpi=300)
plt.show()
```

```python
from scipy.interpolate import make_interp_spline

x = df_merged["pH"]
y = df_merged["Buffer capacity"]

# สร้างค่า x ใหม่ที่ละเอียดขึ้นสำหรับเส้นโค้ง
x_new = np.linspace(x.min(), x.max(), 200)
spl = make_interp_spline(x, y, k=3)
y_smooth = spl(x_new)

plt.figure(figsize=(13, 8))
plt.plot(x_new, y_smooth, color='#0072B5', linewidth=3, label='Smoothed Curve')
#plt.scatter(x, y, color='#FF9900', edgecolor='black', s=90, zorder=5, label='Data Points')
plt.xlabel("pH", fontsize=18, fontweight='bold')
plt.ylabel("Buffer Capacity (mol/L·pH)", fontsize=18, fontweight='bold')
plt.title("Relationship between pH and Buffer Capacity", fontsize=22, fontweight='bold', color='#333333')
plt.grid(True, linestyle=':', alpha=0.6)
plt.tight_layout(pad=3)
plt.xlim(x.min() - 0.5, x.max() + 0.5)
plt.ylim(bottom=0)
plt.legend(fontsize=14, loc='upper right', frameon=True, facecolor='white', edgecolor='gray')
plt.savefig("Spline interpolation_buffer.png", dpi=350)
plt.show()
```
